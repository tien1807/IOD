```python
# insert code here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

### Load data
Load the Diabetes dataset from **SciKit-Learn**.

```python
# insert code here
# Load data
diabetes = load_diabetes()
X, y = diabetes.data, diabetes.target
```

```python
# Print basic information about the dataset
print("Input shape:", X.shape)
print("Output shape:", y.shape)
print("\nFeature names:", diabetes.feature_names)
print("\nSample input:\n", X[:5])
print("\nSample output:\n", y[:5])
```

### Prepare input data

```python
# insert code here
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Split the data (training/test)

```python
# insert code here
# Split the data (training/test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

### Create the model's architecture

```python
# insert code here
model = Sequential([
    Dense(8, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(4, activation='relu'),
    Dense(1, activation='linear')
])
```

```python
print(model.summary())
```

### Compile the model

```python
# insert code here
# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae'])
```

### Fit the training data

```python
%%time
# insert code here
history = model.fit(X_train, y_train, validation_split=0.2, epochs=200, batch_size=32, verbose=1)
```

### Create predictions

```python
# insert code here
y_pred = model.predict(X_test)
```

### Evaluate the model

```python
# insert code here
mse, mae = model.evaluate(X_test, y_test)
print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
```

### Visualisation of cost

```python
# Visualisation of loss and accuracy (MAE)
plt.figure(figsize=(12, 5))
```

```python
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss (MSE) - 8-4-1 Architecture')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
```

```python
from sklearn.metrics import r2_score
# Calculate R-squared
r2 = r2_score(y_test, y_pred)
print(f"R-squared: {r2}")
```

```python
# insert code here
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
# Visualisation of Actual vs Predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values - 8-4-1 Architecture')
plt.show()
```

### Visualisation of residuals

```python
# insert code here
residuals = y_test - y_pred.flatten()
plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuals)
plt.title('Residuals vs Predicted')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```
