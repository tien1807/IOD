## Prepare the text
All the text handling and preparation concerned with the changes and modifications from the raw source text to a format that will be used for the actual processing, things like:
- handle encoding
- handle extraneous and international characters
- handle symbols
- handle metadata and embedded information
- handle repetitions (such multiple spaces or newlines)

Clean text.

```python
def clean_text(text):
    # reduce multiple spaces and newlines to only one
    text = re.sub(r'(\s\s+|\n\n+)', r'\1', text)
    # remove double quotes
    text = re.sub(r'"', '', text)

    return text
```

```python
import unicodedata
```

```python
def clean_text(text):
    # Convert to lowercase
    text = text.lower()
    
    # Handle encoding issues
    text = text.encode('ascii', 'ignore').decode('utf-8')
    
    # Normalize unicode characters
    text = unicodedata.normalize('NFKD', text)
    
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    
    # Remove email addresses
    text = re.sub(r'\S*@\S*\s?', '', text)
    
    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Handle symbols and punctuation
    text = re.sub(r'[^\w\s]', '', text)
    
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Handle repetitions
    text = re.sub(r'(\w)\1{2,}', r'\1\1', text)  # Replace 3 or more repetitions with 2
    
    return text

# Apply the clean_text function to your dataset
df['cleaned_text'] = df['text'].apply(clean_text)
```

## Work the text
Using techniques learned in previous labs, remove StopWords, punctuation, and digits. Entities can be retained. Return the lemmatised form of any remaining words in lower case form.

This removes meaningless information.

```python
# Complete the function
def convert_text(text):
    # Process the text with spaCy
    doc = nlp(text)
    print("text",text)
    print("doc",doc)
    # Initialize an empty list to store processed tokens
    processed_tokens = []
    
    for token in doc:
        # Check if the token is not a stopword, punctuation, or digit
        if not token.is_stop and not token.is_punct and not token.like_num:
            # If it's a named entity, keep the original form
            if token.ent_type_:
                processed_tokens.append(token.text.lower())
            else:
                # Otherwise, use the lemmatized form
                processed_tokens.append(token.lemma_.lower())
    print("processed_tokens",processed_tokens)
    
    # Join the processed tokens back into a string
    return ' '.join(processed_tokens)
```

## Create a Bag-of-Words Model

```python
# create a matrix of word counts from the text
counts = CountVectorizer()
```

```python
# do the actual counting
A = counts.fit_transform(X_train, y_train)
```

```python
# create a classifier using SVC
classifier = SVC(kernel='linear', probability=True)
```

```python
# train the classifier with the training data
classifier.fit(A, y_train)
```

```python
# do the transformation for the test data
# NOTE: use `transform()` instead of `fit_transform()`
B = counts.transform(X_test)
```

```python
# make predictions based on the test data
predictions = classifier.predict(B)

# store probabilities of predictions being 1
probabilities = classifier.predict_proba(B)[:, 1]
```

```python
# check the accuracy
print('Accuracy: %.4f' % accuracy_score(y_test, predictions))
```

## Repeat using TF-IDF

## Defining a helper function to show results and charts

```python

def show_summary_report(actual, prediction, probabilities):

    if isinstance(actual, pd.Series):
        actual = actual.values.astype(int)
    prediction = prediction.astype(int)

    accuracy_ = accuracy_score(actual, prediction)
    precision_ = precision_score(actual, prediction)
    recall_ = recall_score(actual, prediction)
    roc_auc_ = roc_auc_score(actual, probabilities)

    print('Accuracy : %.4f [TP / N] Proportion of predicted labels that match the true labels. Best: 1, Worst: 0' % accuracy_)
    print('Precision: %.4f [TP / (TP + FP)] Not to label a negative sample as positive.        Best: 1, Worst: 0' % precision_)
    print('Recall   : %.4f [TP / (TP + FN)] Find all the positive samples.                     Best: 1, Worst: 0' % recall_)
    print('ROC AUC  : %.4f                                                                     Best: 1, Worst: < 0.5' % roc_auc_)
    print('-' * 107)
    print('TP: True Positives, FP: False Positives, TN: True Negatives, FN: False Negatives, N: Number of samples')

    # Confusion Matrix
    mat = confusion_matrix(actual, prediction)

    # Precision/Recall
    precision, recall, _ = precision_recall_curve(actual, probabilities)
    average_precision = average_precision_score(actual, probabilities)

    # Compute ROC curve and ROC area
    fpr, tpr, _ = roc_curve(actual, probabilities)
    roc_auc = auc(fpr, tpr)


    # plot
    fig, ax = plt.subplots(1, 3, figsize = (18, 6))
    fig.subplots_adjust(left = 0.02, right = 0.98, wspace = 0.2)

    # Confusion Matrix
    sns.heatmap(mat.T, square = True, annot = True, fmt = 'd', cbar = False, cmap = 'Blues', ax = ax[0])

    ax[0].set_title('Confusion Matrix')
    ax[0].set_xlabel('True label')
    ax[0].set_ylabel('Predicted label')

    # Precision/Recall
    step_kwargs = {'step': 'post'}
    ax[1].step(recall, precision, color = 'b', alpha = 0.2, where = 'post')
    ax[1].fill_between(recall, precision, alpha = 0.2, color = 'b', **step_kwargs)
    ax[1].set_ylim([0.0, 1.0])
    ax[1].set_xlim([0.0, 1.0])
    ax[1].set_xlabel('Recall')
    ax[1].set_ylabel('Precision')
    ax[1].set_title('2-class Precision-Recall curve')

    # ROC
    ax[2].plot(fpr, tpr, color = 'darkorange', lw = 2, label = 'ROC curve (AUC = %0.2f)' % roc_auc)
    ax[2].plot([0, 1], [0, 1], color = 'navy', lw = 2, linestyle = '--')
    ax[2].set_xlim([0.0, 1.0])
    ax[2].set_ylim([0.0, 1.0])
    ax[2].set_xlabel('False Positive Rate')
    ax[2].set_ylabel('True Positive Rate')
    ax[2].set_title('Receiver Operating Characteristic')
    ax[2].legend(loc = 'lower right')

    plt.show()

    return (accuracy_, precision_, recall_, roc_auc_)
```

## Repeating it all for comparison
Repeat the whole lot in one big block using the show_summary_report function.

Find 'Accuracy', 'Precision', 'Recall', 'ROC_AUC' using CountVectorizer and TfidfVectorizer and keep the result in a dataframe.

```python
# ANSWER
# CountVectorizer
counts = CountVectorizer()
A_counts = counts.fit_transform(X_train)
classifier_counts = SVC(kernel='linear', probability=True)
classifier_counts.fit(A_counts, y_train)
B_counts = counts.transform(X_test)
predictions_counts = classifier_counts.predict(B_counts)
probabilities_counts = classifier_counts.predict_proba(B_counts)[:, 1]

# TfidfVectorizer
tfidf = TfidfVectorizer()
A_tfidf = tfidf.fit_transform(X_train)
classifier_tfidf = SVC(kernel='linear', probability=True)
classifier_tfidf.fit(A_tfidf, y_train)
B_tfidf = tfidf.transform(X_test)
predictions_tfidf = classifier_tfidf.predict(B_tfidf)
probabilities_tfidf = classifier_tfidf.predict_proba(B_tfidf)[:, 1]

# Compare results
results = pd.DataFrame({
    'Method': ['CountVectorizer', 'TfidfVectorizer'],
    'Accuracy': [
        accuracy_score(y_test, predictions_counts),
        accuracy_score(y_test, predictions_tfidf)
    ],
    'Precision': [
        precision_score(y_test, predictions_counts),
        precision_score(y_test, predictions_tfidf)
    ],
    'Recall': [
        recall_score(y_test, predictions_counts),
        recall_score(y_test, predictions_tfidf)
    ],
    'ROC_AUC': [
        roc_auc_score(y_test, probabilities_counts),
        roc_auc_score(y_test, probabilities_tfidf)
    ]
})

print(results)

# Visualize results
show_summary_report(y_test, predictions_counts, probabilities_counts)
show_summary_report(y_test, predictions_tfidf, probabilities_tfidf)
```
