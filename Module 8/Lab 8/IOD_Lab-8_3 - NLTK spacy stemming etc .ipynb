## Work the data
- if necessary or desired
    - remove text or content, e.g. quotes (") or metadata (===)
    - add content or markers, e.g. (#FLAG, --NAME--)
    - remove or convert special symbols, e.g. "é" to "e"
    - remove or convert emoticons

```python
import re
```

```python
# ANSWER
def clean_text(text):
    # Remove metadata (===)
    text = re.sub(r'={2,}.*?={2,}', '', text)
    
    # Remove quotes
    text = re.sub(r'"', '', text)
    
    # Convert special characters
    text = text.replace('é', 'e')
    
    # Remove newline characters
    text = re.sub(r'\n+', ' ', text)
    
    # Add markers for names (this is a simple example, might need refinement)
    text = re.sub(r'([A-Z][a-z]+ [A-Z][a-z]+)', r'--\1--', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text)
    
    return text.strip()

# Clean the text
cleaned_text = clean_text(text)

# Print the first 1000 characters of the cleaned text
print(cleaned_text[:1000])

# Let's also count the number of name markers we added
name_count = len(re.findall(r'--.*?--', cleaned_text))
print(f"\nNumber of potential names marked: {name_count}")
```

## Helper method

```python
# create a bar chart of the frequency of the words in the text
def plot_words(tokens, top = 30):
    tokens_counter = Counter(tokens)
    tok, val = zip(*tokens_counter.most_common(top))
    
    plt.figure(figsize = (16, 6))
    plt.bar(tok, val)
    plt.title(f'Top {top} most common words (Total unique words: {len(tokens_counter)})')
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.xticks(rotation = 90)
    plt.tight_layout()
    plt.show()

# Now let's use this function with our cleaned text
words = cleaned_text.split()
plot_words(words)
```

### Stemming

```python
## nltk
ps = nltk.porter.PorterStemmer()
stemmed = ' '.join([ps.stem(word) for word in text.split()])
```

### Lemmatisation

```python
## nltk
wl = nltk.stem.WordNetLemmatizer()
lemma = ' '.join([wl.lemmatize(word) for word in text.split()])
```

```python
## nltk
dot = stemmed.find('.') + 1
sl = stemmed[:dot].split()
dot = lemma.find('.') + 1
ll = lemma[:dot].split()

print('i | Stem           Lemma')
print('--| -------------- ------------')
for i, p in enumerate(zip(sl, ll)):
    print('%2d| %-12r   %-12r' % (i+1, p[0], p[1]))
```

```python
plot_words(stemmed.split())
```

```python
plot_words(lemma.split())
```

### Entity recognition

```python
## nltk
entities = nltk.chunk.ne_chunk(tagged)

for e in entities:
    s = re.sub(r'[\(\)]', '', str(e))
    if s.find('/NNP') > 0:
        t = s.split()[0]
        n = ' '.join([re.sub(r'/NNP', '', x) for x in s.split()[1:]])
        print('%-12s: %s' % (t, n))
```
