```python
# IMPORT LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

### 1. Import the data

```python
csv = 'ais.csv'
df = pd.read_csv(csv)
```

### 2. Do some EDA of relationships between features.

### 3. Prepare the data for clustering

1. Remove the `sex`, `sport` column. We will see if the clusters from K-Means end up like the actual sex.
2. Put the features on the same scale.

```python
# ANSWER
data_cluster = df.drop(['sex', 'sport'], axis=1)
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_cluster)
```

```python
data_scaled = pd.DataFrame(data_scaled, columns=data_cluster.columns)
```

### 4. Clustering with K-Means

- Cluster the data to our target groups.
    - Use sklearn.cluster.KMeans
- We know that there are 2 actual classes. Let's try using 2, 4. Run the clustering algorithm with each of those k values.

```python
# ANSWER
# Perform K-Means clustering with k=2
kmeans_2 = KMeans(n_clusters=2, random_state=42, n_init=10)
labels_2 = kmeans_2.fit_predict(data_scaled)

# Perform K-Means clustering with k=4
kmeans_4 = KMeans(n_clusters=4, random_state=42, n_init=10)
labels_4 = kmeans_4.fit_predict(data_scaled)

# Add the cluster labels to the original dataset
df['cluster_2'] = labels_2
df['cluster_4'] = labels_4

# Display the first few rows of the result
print(df[['sex', 'cluster_2', 'cluster_4']].head(10))
```

### 6. Elbow Method For Optimal k

For each k value, we will initialise k-means and use the inertia attribute to identify the sum of squared distances of samples to the nearest cluster centre. As k increases, the sum of squared distance tends to zero.

```python
# ANSWER
# Calculate inertia for different values of k
inertias = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(data_scaled)
    inertias.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertias, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()

# Print the inertia values
for k, inertia in zip(k_range, inertias):
    print(f"k={k}: inertia={inertia}")
```

### 7. Plot `wt` vs. `pcBfat` using the clusters from each model.

- Plot wt, pcBfat using hue sex
- Plot wt, pcBfat using hue label

The labels calculated above can be a new column of the dataset.

```python
# ANSWER
# Function to create scatter plot
def plot_scatter(x, y, hue, data, title):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=x, y=y, hue=hue, data=data)
    plt.title(title)
    plt.show()

# Plot wt vs pcBfat using sex as hue
plot_scatter('wt', 'pcBfat', 'sex', df, 'Weight vs Body Fat % (by Sex)')

# Plot wt vs pcBfat using cluster_2 as hue
plot_scatter('wt', 'pcBfat', 'cluster_2', df, 'Weight vs Body Fat % (K-Means, k=2)')

# Plot wt vs pcBfat using cluster_4 as hue
plot_scatter('wt', 'pcBfat', 'cluster_4', df, 'Weight vs Body Fat % (K-Means, k=4)')

# Additional analysis: confusion matrix for k=2
from sklearn.metrics import confusion_matrix
import numpy as np

# Assuming 'f' is encoded as 0 and 'm' as 1
true_labels = np.where(df['sex'] == 'f', 0, 1)
cm = confusion_matrix(true_labels, df['cluster_2'])

print("Confusion Matrix for k=2:")
print(cm)
```

```python

# Calculate accuracy
accuracy = (cm[0, 0] + cm[1, 1]) / cm.sum()
print(f"Accuracy: {accuracy:.2f}")
```
