```python
import pandas as pd
import numpy as np
import math
import seaborn as sns
import pandas as pd

from scipy import stats
from matplotlib import pyplot as plt
```

```python
### Display summary statistics of the dataset
#Introduction
#This analysis aims to predict Apple's (AAPL) stock closing prices using historical data.
#We will perform data wrangling, exploratory data analysis, feature engineering, and build a linear regression model to achieve this goal.

# Load the dataset
csv = 'HistoricalQuotes.csv'
data = pd.read_csv(csv)
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Display basic information about the dataset
print(data.info())
```

```python
# Display summary statistics of the dataset
data.describe().T
```

```python
##Data Wrangling
#Convert Date to Datetime and Clean Data

# Remove leading/trailing whitespaces from column names
data.columns = data.columns.str.strip()

```

```python

# Convert 'Date' to datetime format
data['Date'] = pd.to_datetime(data['Date'])

```

```python
# Remove the dollar sign and convert to float for relevant columns
for column in ['Close/Last', 'Open', 'High', 'Low']:
    data[column] = data[column].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Rename columns for consistency
data.rename(columns={'Close/Last': 'Close'}, inplace=True)
```

```python
# Set 'Date' as the index
data.set_index('Date', inplace=True)
```

```python
# Display the cleaned data
data.head()
```

```python
##Feature Engineering
#Create Lag Features and Moving Averages
```

```python
# Create lag features
data['Close_Lag1'] = data['Close'].shift(1)
data['Close_Lag2'] = data['Close'].shift(2)
data['Close_Lag3'] = data['Close'].shift(3)
```

```python

# Create moving averages
data['MA_5'] = data['Close'].rolling(window=5).mean()
data['MA_10'] = data['Close'].rolling(window=10).mean()
data['MA_20'] = data['Close'].rolling(window=20).mean()
```

```python
# Create exponential moving averages
data['EMA_5'] = data['Close'].ewm(span=5, adjust=False).mean()
data['EMA_10'] = data['Close'].ewm(span=10, adjust=False).mean()
data['EMA_20'] = data['Close'].ewm(span=20, adjust=False).mean()
```

```python
# Calculate Relative Strength Index (RSI)
def calculate_rsi(data, window):
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

data['RSI_14'] = calculate_rsi(data, 14)
```

```python
# Drop rows with NaN values created by shift and rolling operations
data.dropna(inplace=True)
```

```python
# Display the first few rows with new features
data.head()
```

```python
##Building a Linear Regression Model
#Define Features and Target Variable
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

# Define features and target variable
features = ['Close_Lag1', 'Close_Lag2', 'Close_Lag3', 'MA_5', 'MA_10', 'EMA_5', 'EMA_10', 'EMA_20', 'RSI_14']
target = 'Close'

X = data[features]
y = data[target]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
#Train the Linear Regression Model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```

```python
# Generate predictions
y_pred = model.predict(X_test)

# Combine predictions with actual values
predictions = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}, index=y_test.index)

# Plotting the results
# Plotting the results as a scatter plot
plt.figure(figsize=(14, 7))
plt.scatter(predictions.index, predictions['Actual'], label='Actual Closing Prices', color='green', alpha=0.6)
plt.title('Actual vs. Predicted Closing Prices')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()

plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Close Price', color='green')
plt.title('Apple (AAPL) Stock Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```
